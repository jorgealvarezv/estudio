## Graficas con matplotlib Orientado a objetos

```python

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
y = x**2
```

#### Se crea el grafico 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,0.5,0.5])
ax.plot(x, y, 'r')

```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,0.5,0.5])
ax2 = fig.add_axes([0,0.6,0.5,0.5])
ax.set_facecolor('black')
ax.plot(x, y, 'r')
ax2.plot(y, x, 'b')
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2)
ax[0,0].plot(x,y,'r')
ax[1,0].plot(y,x,'b')
ax[1,1].plot(-y,x,'b')
ax[0,1].plot(-x,y,'r')

ax[0,0].set_facecolor('black')
ax[0,1].set_facecolor('black')
ax[1,0].set_facecolor('black')
ax[1,1].set_facecolor('black')
fig.tight_layout()
```
